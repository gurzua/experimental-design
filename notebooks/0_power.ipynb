# Power (statistical power)

![statistical_poer](images/statistical_power.png)

Statistical power is the probability of correctly detecting a real effect when it exists.

The formula is defined as: **Power = 1- Beta**
- Beta (type error II): The probability of failing to detect a real effect when it exists
- A commonly accepted level of statistical power is 80%


#### Example
- effect size: 0.4
- sample size: 50
- alpha: 5%

```python
from statsmodels.stats.power import TTestIndPower

analysis = TTestIndPower()

power = analysis.power(
    effect_size=0.4,
    nobs1=50,
    alpha=0.05,
    alternative='two-sided'
)

print(f'There is a {round(power,2)*100}% probability of detecting a true effect if it exists.' )
```
