# Testing Normality

## Visual normality in an agricultural experiment
You have been contracted by an agricultural firm conducting an experiment on chickens, divided into four groups, each fed a different diet. Weight measurements were taken every second day for 20 days.

You'll analyze *chicken_data (df)* to assess normality, which will determine the suitability of parametric statistical tests, beginning with a visual examination of the data distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.gofplots import qqplot
from scipy.stats.distributions import norm
```

```python
df = pd.read_csv('chick_weight.csv')
df.head()
```

### Plot the distribution of the chickens' weight using the kernel density estimation (KDE) to visualize normality.

```python
sns.displot(data=df, x='weight', kind='kde')
```

### Create a qq plot with a standard line of the chickens' weight to assess normality visually.

```python
qqplot(df['weight'], line='s', dist=norm)
plt.show()
```

### Subset chicken_data for a 'Time' of 2, and plot the KDE of 'weight' from subset_data to check if data is normal across time.

```python
subset_time2 = df[df['Time']==2]

sns.displot(data=subset_time2, x='weight', kind='kde', fill=True)
plt.show()
```

## Analytical methods
Carrying on from your previous work, your visual inspections of the data indicate it may not be a normal dataset overall, but that the initial time point may be.

Build on your previous work by using analytical methods to determine the normality of the dataset.

### Shapiro-Wilk (good for smaller datasets)
Run a Shapiro-Wilk test of normality on the 'weight' column and print the test statistic and p-value.

```python
from scipy.stats import shapiro

# Run a Shapiro-Wilk normality test on the weight column
stat, p_value = shapiro(df['weight'])

alpha=0.05
print(f"p-value: {round(p_value, 4)}, test stat: {round(stat, 4)}")
if p_value < alpha:
    print('Reject the null hypothesis, the data do not follow a normal distribution')
else:
    print('Fail to reject the null hypothesis')
```

### Anderson-Darling test
Returns an object with the statistic, significance levels and critical values.


In the example, alpha is set to 5%. You then identify its corresponding entry in the list of significance levels, which is 5. At that same index, you obtain the associated critical value. If the test statistic exceeds this critical value, the null hypothesis is rejected, indicating that the data do not follow a normal distribution.

```python
from scipy.stats import anderson

result = anderson(df['weight'], dist='norm')
alpha = 0.05
stat = result.statistic
critical_value = result.critical_values[result.significance_level == 5]

print(f'Significance levels: {result.significance_level}')
print(f'Critical values: {result.critical_values}')
print(f'Test statistic: {round(result.statistic,4)}')
print(f"alpha's critical value (5%) {critical_value}")


if stat < critical_value:
    print('Fail to reject the null hypothesis')
else:
    print('Reject the null hypothesis, the data do not follow a normal distribution')
```
