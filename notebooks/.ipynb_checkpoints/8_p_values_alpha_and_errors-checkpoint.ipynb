```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
```

```python
investment_data = pd.read_csv('../data/investment_returns.csv')
investment_data.head()
```

```python
sns.displot(data=investment_data, x='Annual_Return', hue='Strategy_Type', kind='kde')
plt.show()
```

```python
quantitative = investment_data.loc[investment_data['Strategy_Type']=='Quantitative', 'Annual_Return']
fundamental = investment_data.loc[investment_data['Strategy_Type']=='Fundamental', 'Annual_Return']

t_stat, p_val = ttest_ind(quantitative, fundamental)
print(p_val)
```

The P-value suggests that there's a statistically significant difference in annual return between 'Quantitative' and 'Fundamental' strategies, assuming an alpha of 0.05
