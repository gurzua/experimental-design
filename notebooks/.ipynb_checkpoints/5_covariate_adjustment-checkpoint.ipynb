# Covariate adjustment with chick growth
Imagine studying in agricultural science the growth patterns of chicks under various dietary regimens. The data from this study sheds light on the intricate relationship between their respective diets and the consequent impact on their weight. This data includes weight measurements of chicks at different ages, allowing for an exploration of covariate adjustment. Time serves as a covariate, potentially influencing the outcome variable: the weight of the chicks.

```python
import pandas as pd
import numpy as np
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
chick_data = pd.read_csv('../data/chick_weight.csv')
chick_data.head(3)
```

Ancova test without interaction
- Time: covariant
- Diet: Factor (Categorical)
- Assumes the same slot for every diet

```python
# Perform ANCOVA with Diet and Time as predictors
model = ols('weight ~ C(Diet) + Time', data=chick_data).fit()

# Print a summary of the model
print(model.summary())
```

#### Model summary
- R-squared: 0.745, meaning that 74% of the variability in weight is explained by Time and Diet
- F-statistic: 419.2, indicating that the overall model is highly significat, at least one predictor has real effect over weight

#### Coefficient interpretation
- Intercept: 10.2 (P>|t|=0.001), this means that when Time=0 and Diet=1, the expected weight is 10.92
- Time: 8.7505 (P>|t|=0.000), this means that for each one-unit increase in time, weight increases by approximately 8.75 units.

#### Diet interpretation
Each coefficient "C(Diet)" represents the difference in mean weight relative to Diet=1

| Comparation | Difference |
|------------ |------------|
| Diet 2 vs 1 | +16.1661   |
| Diet 3 vs 1 | +36.4994   |
| Diet 4 vs 1 | +30.2335   |

#### Conclusion:
After adjusting for time, there are statistically significant differences in mean weight across diets.
Diet 3 shows the highest mean weight, followed by Diet 4 and Diet 2, all of which are significantly higher than Diet 1.
Weight increases significantly over time, assuming a common growth rate across diets.

```python
# Visualize Diet effects with Time adjustment
sns.lmplot(x='Time', y='weight', 
         hue='Diet', 
         data=chick_data)
plt.show()
```

#### Conclusion:
Weight increases over time across all diets, but not all diets are equally effective. Diet 3 (and to a lesser extent Diet 4) promotes a faster weight gain, and these differences become more pronounced as time progresses.
